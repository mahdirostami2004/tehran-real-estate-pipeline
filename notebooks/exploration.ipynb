# Tehran Real Estate — Exploratory Data Analysis

Quick EDA on the cleaned dataset produced by the ETL pipeline.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT))

import pandas as pd
from src.extract import extract
from src.transform import transform
```

```python
raw = extract(include_incoming=False)
clean = transform(raw)
print(f"Raw rows: {len(raw):,}")
print(f"Clean rows: {len(clean):,}")
clean.head()
```

```python
clean.describe()
```

```python
by_address = (
    clean.groupby("address")
    .agg(
        listings=("record_hash", "count"),
        avg_price_per_sqm=("price_per_sqm", "mean"),
        avg_area=("area_sqm", "mean"),
    )
    .sort_values("avg_price_per_sqm", ascending=False)
)
by_address.head(10)
```

```python
by_rooms = clean.groupby("rooms").agg(
    count=("record_hash", "count"),
    avg_price_toman=("price_toman", "mean"),
    avg_price_per_sqm=("price_per_sqm", "mean"),
)
by_rooms
```

```python
amenities = clean[["parking", "warehouse", "elevator", "price_per_sqm"]]
amenities.groupby(["parking", "elevator"])["price_per_sqm"].mean().round(0)
```
